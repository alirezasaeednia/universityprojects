```python
    import matplotlib.pyplot as plt
    encoded = []
    current_level = 1  
    binary_string='0101010101'
    print(encoded)
```

```python
for bit in binary_string:
        if bit == '0':
            encoded.append(current_level)  
        else:
            current_level = -current_level 
            encoded.append(current_level) 
print(encoded)
              
```

```python
    x = range(len(binary_string)+1)
    y = [1] + encoded 
    
    print(x)   
    print(y)
```

```python
    %matplotlib inline 
    plt.step(x, y,color='g')
    
    for i, bit in enumerate(binary_string):
        plt.text(i+0.5, encoded[i], bit, ha='center', va='bottom',color='r')
    
    plt.xlabel('Time')
    plt.ylabel('Voltage')
    plt.title('NRZI')
    plt.grid(True)
    plt.show()
```
