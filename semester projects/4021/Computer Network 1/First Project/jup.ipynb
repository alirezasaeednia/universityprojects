```python

def plot_nrz_pulse(binary_string):
        %matplotlib inline 
        time = range(0, len(binary_string)+1)
        signal = [1 if bit == '1' else 0 for bit in binary_string]
        signal_update=signal+[0]
        plt.step(time, signal_update, where='post',color='b')
        plt.xlabel('Time')
        plt.ylabel('Voltage Level')
        plt.title('nrz : ')
        plt.show()
    
    
plot_nrz_pulse('10001') 
 
```

```python

def plot_nrz_pulse(binary_string):
        %matplotlib inline 
        time = range(0, len(binary_string)+1)
        signal = [1 if bit == '1' else 0 for bit in binary_string]
        signal_update=signal+[0]
        plt.step(time, signal_update, where='post',color='r')
        for i,bit in enumerate(binary_string):
                plt.text(i + 0.5, 1.08, bit, ha='center') 
        plt.xlabel('Time')
        plt.ylabel('Voltage Level')
        plt.show()
    
    
plot_nrz_pulse('10001') 
 
```
